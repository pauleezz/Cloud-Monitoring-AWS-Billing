```python
import pandas as pd
import numpy as np
import seaborn as sns
import glob
import os
%matplotlib inline
```

```python
df = pd.read_csv(os.path.join("../295273672130-cur-with-tags-2021-02.csv"))

df.head()
```

```python
# drop the unique value
df = df.drop(columns=['InvoiceID', 'PayerAccountId', 'LinkedAccountId', 'RecordId', 'ReservedInstance', 'user_Division', 'user_ENV','UnblendedRate'])
# drop ItemDescription for already exist feature
df = df.drop(columns=['ItemDescription'])

df.head(10)
```

```python
createby_df = df.dropna(subset=['aws_createdBy'])
iam_df = createby_df[createby_df['aws_createdBy'].str.contains('IAMUser')].copy()
iam_df['aws_createdBy'] = iam_df['aws_createdBy'].apply(lambda x: x[30:])

# iam_df
iam_df['aws_createdBy'].unique()
#使用頻率高、使用服務user
```

```python
iam_df.groupby(by='ProductName').count()
```

```python
iam_df = iam_df.reset_index(drop=True)
temp = iam_df.copy()
num = iam_df.shape[0]
time = 1
arr =[]
place=[0]
for n in range(num):
    try:
        if n==num-1:
            arr.append(1)
            break
        nex = n+1
        nex2 = n+2
        first = iam_df.iloc[n:nex]['UsageEndDate'].squeeze()
        second = iam_df.iloc[nex:nex2]['UsageStartDate'].squeeze()
        if first == second:
            place.append(n)
            start_index = place[1]
            time += 1
            temp = temp.drop(nex)
        else:
            
            arr.append(time)
            if time > 1:
                temp.loc[start_index,'UsageEndDate']=iam_df.iloc[start_index+time-1,:]['UsageStartDate'] # set the new UsageEndDate
            time = 1
            place=[]
            place.append(n)
    except:
        print(n)
temp['period(h)']=arr   
```

```python
iam_df = temp.copy()
```
