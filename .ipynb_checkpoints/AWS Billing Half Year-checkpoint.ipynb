```python
createby_df = df.dropna(subset=['aws_createdBy'])
iam_df = createby_df[createby_df['aws_createdBy'].str.contains('IAMUser')].copy()
iam_df['aws_createdBy'] = iam_df['aws_createdBy'].apply(lambda x: x[30:])

# iam_df
iam_df['aws_createdBy'].unique()
#使用頻率高、使用服務user
```

```python
iam_df.groupby(by='ProductName').count()
```

```python
iam_df = iam_df.reset_index(drop=True)
temp = iam_df.copy()
num = iam_df.shape[0]
time = 1
arr =[]
dic={}
for n in range(num):
    if n==num-1:
        arr.append(1)
        break
    nex = n+1
    nex2 = n+2
    first = iam_df.iloc[n:nex]['UsageEndDate'].squeeze()
    second = iam_df.iloc[nex:nex2]['UsageStartDate'].squeeze()
    if first == second:
        time += 1
        temp = temp.drop(nex)
    else:
        arr.append(time)
        time = 1
temp['period(h)']=arr
```

```python
iam_df = temp.copy()
```

```python
iam_col = iam_df.columns.tolist()
iam_col.remove('UsageStartDate')
iam_col.remove('UsageEndDate')
iam_col.remove('aws_createdBy')
iam_col.remove('period(h)')
```

```python
iam_fre_df = iam_df.copy()
def fe_fun(col):
    for i in col:
        fe = iam_fre_df.groupby(i).size()/len(iam_fre_df)
        iam_fre_df.loc[:, i+'_freq_encode'] = iam_fre_df[i].map(fe)

fe_fun(iam_col)
```

```python
iam_fre_df = iam_fre_df.drop(columns=['ProductName', 'UsageType', 'Operation', 'AvailabilityZone', 'UsageStartDate', 'UsageEndDate','ResourceId'])
```

```python
iam_fre_df = iam_fre_df.dropna()
```

```python
iam_fre_df
```

```python
iam_freq_col = iam_fre_df.columns.tolist()
iam_freq_col = iam_freq_col[:2] + iam_freq_col[3:-1]+ [iam_freq_col[2]]
iam_freq_col
```

```python
iam_fre_df = iam_fre_df[iam_freq_col]
iam_fre_df
```

```python
from sklearn.cluster import KMeans
import numpy as np
kmeans = KMeans(n_clusters=4, random_state=0).fit(iam_fre_df.iloc[: , :-1])
user_kmeans = kmeans.labels_
iam_fre_kmeans_df = iam_fre_df.assign(kmeans = user_kmeans)
iam_fre_kmeans_df['kmeans_num'] = iam_fre_kmeans_df['kmeans']
iam_fre_kmeans_df.iloc[:, -3:].groupby(by=['aws_createdBy', 'kmeans']).count()
```
